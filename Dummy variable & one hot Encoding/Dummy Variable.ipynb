```python
data = {
    'area': [2600, 3000, 3200, 3600, 4000, 2600, 2800, 3300, 3600, 2600],
    'town': ['monroe township', 'monroe township', 'monroe township', 'monroe township', 'monroe township',
             'west windsor', 'west windsor', 'west windsor', 'west windsor', 'robinsville'],
    'price': [550000, 565000, 610000, 595000, 760000, 620000, 650000, 680000, 730000, 600000]
}
df = pd.DataFrame(data)

# Generate dummies and convert to integers
dummies = pd.get_dummies(df.town).astype(int)

# Integrate dummies with the original DataFrame and drop the original 'town' column
df = pd.concat([df, dummies], axis=1).drop('town', axis=1)

# Fit the Linear Regression model
model = LinearRegression()
model.fit(df[['area', 'monroe township', 'robinsville']], df.price)

# Display the final DataFrame
print(df)

# Make predictions
new_data = pd.DataFrame({
    'area': [2500],
    'monroe township': [0],
    'robinsville': [1]
})

predicted_price = model.predict(new_data)

# Display the predicted price
print(predicted_price)
```
